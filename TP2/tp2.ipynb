Exercice 2 — Suppression des doublons

```python
# Identifier les doublons
print(df[df.duplicated()].head())  # Affiche les doublons, sans les originaux
df[df.duplicated()]

print(df[df.duplicated(keep=False)].head())  # Inclut les originaux

# Supprimer les doublons
df = df.drop_duplicates()
print("Nombre de doublons restants :", df.duplicated().sum())
```

Exercice 3

```python
# Vérifiez le type de Start Year
print(df['Start Year'].dtypes)
# Affichez quelques valeurs problématiques (contenant "Year" ou "AD")
valpb = df["Start Year"].astype(str).str.contains("Year|AD", na=False)

print(df.loc[valpb, "Start Year"].head(10))
# Nettoyez la colonne : supprimez le texte et convertissez en numérique
df["Start Year"] = df["Start Year"].str.replace("Year", "", regex=False)
df["Start Year"] = df["Start Year"].str.replace("AD", "", regex=False)
df['Start Year'] = pd.to_numeric(df['Start Year'], errors='coerce')

# Vérifiez le résultat
print(df['Start Year'].dtypes)
```

Exercice 4 

```python
# Pour Region (peu de nulls) : supprimez les lignes avec null
df.dropna(subset=['Region'])          
# Pour Event Name : remplacez les nulls par "Non nommé"
df['Event Name'].fillna('Non nommé')
# Pour Total Deaths : décidez d'une stratégie et appliquez-la
df['Total Deaths'].fillna('Non dit')
# Pour Magnitude : laissez les nulls (n'a pas de sens pour tous les types)
# Vérifiez le résultat
print(df['Region'].isnull().sum())
print(df['Event Name'].isnull().sum())
print(df['Total Deaths'].isnull().sum())
print(df['Magnitude'].isnull().sum())
```

Exercice 5 — Traitement des outliers

```python
# Identifiez les valeurs négatives dans Total Deaths
df[df['Total Deaths'] < 0]
# Identifiez les valeurs aberrantes dans Magnitude (> 10)
df[df['Magnitude'] > 10]
# Décidez : supprimer ou corriger ?
# Appliquez le traitement
df = df[df['Magnitude'] >= 0]
# Vérifiez le résultat
print(df[df['Magnitude'] > 10])
```

Exercice 6

```python
df['Region'] = df['Region'].str.strip().str.lower()
df['Region'].value_counts()
df['Disaster Type']= df["Disaster Type"].str.strip().str.lower()
df['Disaster Type'].value_counts()

df['Country']= df['Country'].str.title().str.strip()
df['Country'].replace({'USA': 'United States Of America'})
df['Country'].value_counts()
```

Exercice 7

```python
# Créez une colonne Decennie à partir de Start Year
df["Decennie"] = (df["Start Year"] // 10) * 10

# Créez une colonne Has_Deaths (booléen : True si Total Deaths > 0)
df["Has_Deaths"] = df["Total Deaths"] > 0

# Vérifiez vos nouvelles colonnes
print(df[["Start Year", "Decennie", "Total Deaths", "Has_Deaths"]].head(10))
print(df.dtypes)
print(df["Has_Deaths"].value_counts())
```

Exercice 8

```python
# Sélectionnez les colonnes utiles pour l'analyse
df_export = df[["DisNo.","Event Name", "Start Year", "Region","Subregion", "Country", "Disaster Type", "Total Deaths", "Magnitude", "Decennie", "Has_Deaths"]]
# Renommez les colonnes si nécessaire (noms clairs, sans espaces)
df_export = df_export.rename(columns={'Event Name' : 'Event_Name','Start Year': 'Start_Year','Disaster Type' : 'Disaster_Type', 'Total Deaths': 'Total_Deaths'})
# Faites une vérification finale avec info() et head()
print(df_export.info())
print(df_export.head())
# Exportez en CSV : catnat_clean.csv
df_export.to_csv("catnat_clean.csv", index=False)
```
